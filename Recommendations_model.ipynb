```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('final_movies.csv')
```

```python
print(data.shape)
data.head()
```

```python
movie_features = data[['IMDB_Score','Genre','Year','Runtime','AvgRating']]
```

```python
movie_features.head()
```

```python
genrefearutes = movie_features.Genre.str.get_dummies(sep='|')
```

```python
movie_final = pd.concat([genrefearutes,movie_features.IMDB_Score,movie_features.Year,movie_features.Runtime,movie_features.AvgRating],axis=1)
```

```python
movie_final
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
movies_final_features = min_max_scaler.fit_transform(movie_final)
```

```python
np.round(movies_final_features,2)
```

```python
movies_final_features.shape
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nbrs = NearestNeighbors(n_neighbors=4, algorithm='brute').fit(movies_final_features)
```

```python
distances, indices = nbrs.kneighbors(movies_final_features)
```

```python
import sqlite3
```

```python
con = sqlite3.connect("db.sqlite3", isolation_level=None,
                           detect_types=sqlite3.PARSE_COLNAMES)
```

```python
cursor = con.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
Movies = pd.read_sql_query("SELECT * FROM Movies_movie", con)
```

```python
Movies[Movies.Title.str.contains("Batman")]
```

```python
Movies
```

```python
Movie_genres = pd.read_sql_query('select * from Movies_movie_Genres',con)
```

```python
Gneres = pd.read_sql_query('select * from Movies_genre',con)
```

```python
Genres = Gneres.set_index('id').to_dict()
```

```python
Movie_genres.genre_id = Movie_genres.genre_id.apply(lambda x: Genres['genre'][x])
```

```python
Movie_genres
```

```python
Movie_genres = Movie_genres.groupby('movie_id')['genre_id'].apply('|'.join).reset_index()
```

```python
genre_features = Movie_genres.genre_id.str.get_dummies(sep='|')
```

```python
Movie_genres = pd.concat([Movie_genres.drop('genre_id', axis=1), genre_features], axis=1)
```

```python
Movie_genres
```

```python
Movie_feartures = Movies[['id','Title', 'Imdb_rating', 'RunTime','Rating','Year']]
```

```python
Movie_feartures = pd.concat([Movie_feartures.set_index('id'), Movie_genres.set_index('movie_id')], axis=1, join='inner')
```

```python
Movie_feartures
```

```python
min_max_scaler = MinMaxScaler()
movies_final_features = min_max_scaler.fit_transform(Movie_feartures.drop(['Title','movie_id'], axis=1 ))
```

```python
movies_final_features.shape
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nbrs = NearestNeighbors(n_neighbors=4, algorithm='brute').fit(movies_final_features)
distances, indices = nbrs.kneighbors(movies_final_features)
```
